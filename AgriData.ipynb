```python
#Create table query
create_table_query = """ CREATE TABLE IF NOT EXISTS agritable(
    Dist_Code INT,
    Year INT,
    State_Code INT,
    State_Name VARCHAR(100),
    Dist_Name VARCHAR(100),
    RICE_AREA_1000_ha FLOAT,
    RICE_PRODUCTION_1000_tons FLOAT,
    RICE_YIELD_Kg_per_ha FLOAT,
    WHEAT_AREA_1000_ha FLOAT,
    WHEAT_PRODUCTION_1000_tons FLOAT,
    WHEAT_YIELD_Kg_per_ha FLOAT,
    MAIZE_AREA_1000_ha FLOAT,
    MAIZE_PRODUCTION_1000_tons FLOAT,
    MAIZE_YIELD_Kg_per_ha FLOAT,
    COTTON_AREA_1000_ha FLOAT,
    COTTON_PRODUCTION_1000_tons FLOAT,
    COTTON_YIELD_Kg_per_ha FLOAT,
    GROUNDNUT_AREA_1000_ha FLOAT,
    GROUNDNUT_PRODUCTION_1000_tons FLOAT,
    GROUNDNUT_YIELD_Kg_per_ha FLOAT,
    OILSEEDS_AREA_1000_ha FLOAT,
    OILSEEDS_PRODUCTION_1000_tons FLOAT,
    OILSEEDS_YIELD_Kg_per_ha FLOAT,
    SUNFLOWER_AREA_1000_ha FLOAT,
    SUNFLOWER_PRODUCTION_1000_tons FLOAT,
    SUNFLOWER_YIELD_Kg_per_ha FLOAT,
    SUGARCANE_AREA_1000_ha FLOAT,
    SUGARCANE_PRODUCTION_1000_tons FLOAT,
    SUGARCANE_YIELD_Kg_per_ha FLOAT
);
"""
```

```python
mycursor.execute(create_table_query)
print("Table created sucessfully")
```

```python
#Insert data
insert_query = """
INSERT INTO agritable (
    Dist_Code, Year, State_Code, State_Name, Dist_Name,
    RICE_AREA_1000_ha, RICE_PRODUCTION_1000_tons, RICE_YIELD_Kg_per_ha,
    WHEAT_AREA_1000_ha, WHEAT_PRODUCTION_1000_tons, WHEAT_YIELD_Kg_per_ha,
    MAIZE_AREA_1000_ha, MAIZE_PRODUCTION_1000_tons, MAIZE_YIELD_Kg_per_ha,
    COTTON_AREA_1000_ha, COTTON_PRODUCTION_1000_tons, COTTON_YIELD_Kg_per_ha,
    GROUNDNUT_AREA_1000_ha, GROUNDNUT_PRODUCTION_1000_tons, GROUNDNUT_YIELD_Kg_per_ha,
    OILSEEDS_AREA_1000_ha, OILSEEDS_PRODUCTION_1000_tons, OILSEEDS_YIELD_Kg_per_ha,
    SUNFLOWER_AREA_1000_ha, SUNFLOWER_PRODUCTION_1000_tons, SUNFLOWER_YIELD_Kg_per_ha,
    SUGARCANE_AREA_1000_ha, SUGARCANE_PRODUCTION_1000_tons, SUGARCANE_YIELD_Kg_per_ha
) VALUES (
    %s, %s, %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s,
    %s, %s, %s
)
"""
```

```python
for index, row in data_frame.iterrows():
    mycursor.execute(insert_query, (
        row['Dist_Code'],
        row['Year'],
        row['State_Code'],
        row['State_Name'],
        row['Dist_Name'],
        row['RICE_AREA_1000_ha'],
        row['RICE_PRODUCTION_1000_tons'],
        row['RICE_YIELD_Kg_per_ha'],
        row['WHEAT_AREA_1000_ha'],
        row['WHEAT_PRODUCTION_1000_tons'],
        row['WHEAT_YIELD_Kg_per_ha'],
        row['MAIZE_AREA_1000_ha'],
        row['MAIZE_PRODUCTION_1000_tons'],
        row['MAIZE_YIELD_Kg_per_ha'],
        row['COTTON_AREA_1000_ha'],
        row['COTTON_PRODUCTION_1000_tons'],
        row['COTTON_YIELD_Kg_per_ha'],
        row['GROUNDNUT_AREA_1000_ha'],
        row['GROUNDNUT_PRODUCTION_1000_tons'],
        row['GROUNDNUT_YIELD_Kg_per_ha'],
        row['OILSEEDS_AREA_1000_ha'],
        row['OILSEEDS_PRODUCTION_1000_tons'],
        row['OILSEEDS_YIELD_Kg_per_ha'],
        row['SUNFLOWER_AREA_1000_ha'],
        row['SUNFLOWER_PRODUCTION_1000_tons'],
        row['SUNFLOWER_YIELD_Kg_per_ha'],
        row['SUGARCANE_AREA_1000_ha'],
        row['SUGARCANE_PRODUCTION_1000_tons'],
        row['SUGARCANE_YIELD_Kg_per_ha'],
    ))

```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
#Load data
df = pd.read_csv(r"C:\Users\BaBuReDdI\Desktop\Data Science\Project\ICRISAT-District Level Data - ICRISAT-District Level Data.csv")
```

```python
#Clean Column Names
df.columns = (
    df.columns
    .str.strip()
    .str.upper()
    .str.replace(" ", "_")
    .str.replace(r"\(1000_HA\)", "1000_HA", regex=True)
    .str.replace(r"\(1000_TONS\)", "1000_TONS", regex=True)
    .str.replace(r"\(KG_PER_HA\)", "KG_PER_HA", regex=True)
    .str.replace("[^A-Z0-9_]", "", regex=True)
)
```

```python
print("COLUMNS CLEANED SUCCESSFULLY!")
print(df.columns)
```

```python
#Top 7 RICE PRODUCTION State Data(Bar_plot)
rice_top7 = df.groupby("STATE_NAME")["RICE_PRODUCTION_1000_TONS"].sum().nlargest(7)

plt.figure(figsize=(10,5))
sns.barplot(x=rice_top7.index, y=rice_top7.values)
plt.title("Top 7 Rice Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
wheat_top5 = df.groupby("STATE_NAME")["WHEAT_PRODUCTION_1000_TONS"].sum().nlargest(5)
```

```python
#Bar Chart
plt.figure(figsize=(10,5))
sns.barplot(x=wheat_top5.index, y=wheat_top5.values)
plt.title("Top 5 Wheat Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(7,7))
plt.pie(wheat_top5.values, labels=wheat_top5.index, autopct="%1.1f%%")
plt.title("Wheat Production Share (%) – Top 5 States")
plt.show()
```

```python
#Oil seed production by top 5 states
oil_top5 = df.groupby("STATE_NAME")["OILSEEDS_PRODUCTION_1000_TONS"].sum().nlargest(5)

plt.figure(figsize=(10,5))
sns.barplot(x=oil_top5.index, y=oil_top5.values, palette="viridis")
plt.title("Top 5 Oilseed Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Top 7 SUNFLOWER PRODUCTION  State
sun_top7 = df.groupby("STATE_NAME")["SUNFLOWER_PRODUCTION_1000_TONS"].sum().nlargest(7)

plt.figure(figsize=(10,5))
sns.barplot(x=sun_top7.index, y=sun_top7.values)
plt.title("Top 7 Sunflower Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
sugar = df.groupby("YEAR")["SUGARCANE_PRODUCTION_1000_TONS"].sum().reset_index()

fig = px.line(sugar, x="YEAR", y="SUGARCANE_PRODUCTION_1000_TONS",
              title="India’s Sugarcane Production (Last 50 Years)")
fig.show()
```

```python
#Rice Production Vs Wheat Production (Last 50y)
rice_year = df.groupby("YEAR")["RICE_PRODUCTION_1000_TONS"].sum()
wheat_year = df.groupby("YEAR")["WHEAT_PRODUCTION_1000_TONS"].sum()

plt.figure(figsize=(12,6))
plt.plot(rice_year.index, rice_year.values, label="Rice", linewidth=3)
plt.plot(wheat_year.index, wheat_year.values, label="Wheat", linewidth=3)
plt.title("Rice vs Wheat Production (50 Years)")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.legend()
plt.show()
```

```python
#Rice Production By West Bengal Districts
wb = df[df["STATE_NAME"] == "West Bengal"]

wb_rice = wb.groupby("DIST_NAME")["RICE_PRODUCTION_1000_TONS"].sum().nlargest(10)

plt.figure(figsize=(10,6))
sns.barplot(x=wb_rice.values, y=wb_rice.index)
plt.title("Rice Production – Top Districts in West Bengal")
plt.xlabel("Production (1000 tons)")
plt.ylabel("District Name")
plt.show()
```

```python
#Top 10 Wheat Production Years From UP
up = df[df["STATE_NAME"] == "Uttar Pradesh"]

up_wheat = up.groupby("YEAR")["WHEAT_PRODUCTION_1000_TONS"].sum().nlargest(10)

plt.figure(figsize=(10,5))
sns.barplot(x=up_wheat.index, y=up_wheat.values)
plt.title("Top 10 Wheat Production Years – UP")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Millet Production (Last 50y)
millet = df.groupby("YEAR")["PEARL_MILLET_PRODUCTION_1000_TONS"].sum()

plt.figure(figsize=(12,6))
plt.plot(millet.index, millet.values)
plt.title("Millet Production Over 50 Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Sorghum Production (Kharif and Rabi) by Region
sorghum_region = (
    df.groupby("STATE_NAME")[["KHARIF_SORGHUM_PRODUCTION_1000_TONS",
                              "RABI_SORGHUM_PRODUCTION_1000_TONS"]]
    .sum()
)
```

```python
sorghum_region.head()
```

```python
sorghum_region.plot(kind="bar", figsize=(12,6))
plt.title("Sorghum Production by Region")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 7 States for Groundnut Production
ground_top7 = df.groupby("STATE_NAME")["GROUNDNUT_PRODUCTION_1000_TONS"].sum().nlargest(7)

plt.figure(figsize=(10,5))
sns.barplot(x=ground_top7.index, y=ground_top7.values)
plt.title("Top 7 Groundnut Producing States")
plt.xticks(rotation=45)
plt.ylabel("Production (1000 tons)")
plt.show()
```

```python
#Soybean Production by Top 5 States and Yield Efficiency
soy = df.groupby("STATE_NAME")[["SOYABEAN_PRODUCTION_1000_TONS", "SOYABEAN_YIELD_KG_PER_HA"]].sum()

soy_top5 = soy.nlargest(5, "SOYABEAN_PRODUCTION_1000_TONS")

fig = px.scatter(
    soy_top5,
    x="SOYABEAN_YIELD_KG_PER_HA",
    y="SOYABEAN_PRODUCTION_1000_TONS",
    text=soy_top5.index,
    title="Soybean: Production vs Yield Efficiency (Top 5 States)",
    size="SOYABEAN_PRODUCTION_1000_TONS"
)
fig.update_traces(textposition="top center")
fig.show()
```

```python
#Oilseed Production in Major States
fig = px.bar(
    oil_top5,
    x=oil_top5.index,
    y=oil_top5.values,
    title="Oilseed Production – Major States",
    labels={"y": "Production (1000 tons)"}
)
fig.show()
```

```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
impact = df[[
    "RICE_AREA_1000_HA", "RICE_PRODUCTION_1000_TONS",
    "WHEAT_AREA_1000_HA", "WHEAT_PRODUCTION_1000_TONS",
    "MAIZE_AREA_1000_HA", "MAIZE_PRODUCTION_1000_TONS"
]]
```

```python
sns.pairplot(impact)
plt.suptitle("Impact of Area on Production – Rice, Wheat, Maize", y=1.02)
plt.show()
```

```python
#Rice vs. Wheat Yield Across States
yield_comp = df.groupby("STATE_NAME")[["RICE_YIELD_KG_PER_HA", "WHEAT_YIELD_KG_PER_HA"]].mean()

yield_comp.plot(kind="bar", figsize=(14,6))
plt.title("Rice vs Wheat Yield Across States")
plt.ylabel("Yield (kg/ha)")
plt.xticks(rotation=90)
plt.show()
```
